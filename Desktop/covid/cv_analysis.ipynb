```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('state_level_latest.csv')
df.drop_duplicates(inplace=True)
new_df1=df.drop(df.columns[[5,8,9,10,11]],axis=1)
new_df1
```

```python

new_df=new_df1.drop(0)
new_df2=new_df.drop(36)
new_df=new_df2.drop(37)
new_df
```

```python
mylist=list(new_df['Confirmed'])
total=[]
v=0
for i in range(len(mylist)):
    v=v + mylist[i]
    total.append(v)




mylist1= list(new_df['Recovered'])
total1=[]
v1=0
for i in range(len(mylist1)):
    v1=v1 + mylist1[i]
    total1.append(v1)

mylist2= list(new_df['Deaths'])
total2=[]
v2=0
for i in range(len(mylist2)):
    v2=v2 + mylist2[i]
    total2.append(v2)
```

```python
new_df['Total_Confirm_Addon'] = total
new_df['Total_Recovered_Addon'] = total1
new_df['Total_Death_Addon'] = total2
new_df= new_df.drop(new_df.columns[[9]],axis=1)
new_df
```

```python
new_df.columns
```

```python

y=[]
c=0
for i in range(len(new_df['Total_Confirm_Addon'])):
    c=i
    y.append(c)
```

```python
plt.figure(figsize=(9,7))
plt.plot(new_df['Total_Confirm_Addon'],y,label="Total Reported",color="red")
plt.plot(new_df['Total_Recovered_Addon'],y,label="Total Recovered", color="blue")
plt.xlabel('No Of Cases (in millions)')
plt.title('Covid Analysis Of 2020 (upto Sep)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(new_df['Total_Death_Addon'],y,label="Total Death", color="red")
plt.xlabel('No Of Cases (in Thousands)')
plt.title('Total no of Deaths')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(18,12))
sns.barplot(x='State_code', y='Confirmed', data=new_df)
plt.title('State Wise Analysis (Most no of Patients)')
plt.xlabel('Name Of State')
plt.ylabel('No of Cases')
plt.show()
```

```python
plt.figure(figsize=(18,12))
sns.barplot(x='State_code', y='Active', data=new_df)
plt.title('Most No of Currently affected Patients')
plt.xlabel('Name Of State')
plt.ylabel('No of Cases')
plt.show()
```

```python
plt.figure(figsize=(18,12))
sns.barplot(x='State_code', y='Deaths', data=new_df)
plt.title('Most No of Deaths')
plt.xlabel('Name Of State')
plt.ylabel('No of Cases')
plt.show()
```

```python
ts=pd.read_csv('patients_data1.csv')
ts.drop_duplicates(inplace=True)
ts.dropna(subset=['Gender'],inplace=True)
ts.dropna(subset=['Age Bracket'],inplace=True)
new_ts= ts.drop(ts.columns[[1,2,3,10,11,12,13,15,16,17,18,19,20,21]],axis=1)

```

```python
ns = new_ts[new_ts['Detected State']=='Kerala']
```

```python
ns = new_ts[new_ts['Detected State']=='Kerala']

plt.figure(figsize=(8,6))
sns.countplot(x='Gender',data=ns, palette='Set1',edgecolor="black")
plt.title('Gender Wise Analysis Of Patients in Kerala')
plt.ylabel('No of cases (in Thousands)')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x='Detected District',data=ns, palette='Set1',edgecolor="black",order=ns['Detected District'].value_counts().index)
plt.title('District Wise Analysis Of Patients in Kerala')
plt.xlabel('Name of District')
plt.ylabel('No of cases (in Hundreds)')
plt.show()
```

```python


age= list(ns['Age Bracket'])
age_mod =[]
sm=0

for i in range(len(age)):
    sm= float(age[i])
    age_mod.append(sm)
age1=[]
sm1=0

for i in range(len(age_mod)):
    sm1= int(age_mod[i])
    age1.append(sm1)
```

```python
ns['Age']= age1
ns1=ns.drop(ns.columns[[1]],axis=1)
```

```python
plt.figure(figsize=(16,8))
sns.scatterplot(x="Age",y="Detected District",data=ns1 , hue="Gender")
plt.title('Age Wise Analysis Of Patients in Kerala')
plt.xlabel('Age')
plt.ylabel('Name Of District')
plt.grid()
plt.show()
```
